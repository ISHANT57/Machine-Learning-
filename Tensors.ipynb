```python
import numpy as np
```

```python
scalar = np.array(5)

print(scalar)
print("Shape:", scalar.shape)
print("Dimensions:", scalar.ndim)
```

```python
vector = np.array([2, 4, 6, 8])

print(vector)
print("Shape:", vector.shape)
print("Dimensions:", vector.ndim)
```

```python
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(matrix)
print("Shape:", matrix.shape)
print("Dimensions:", matrix.ndim)
```

```python
tensor_3d = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])

print(tensor_3d)
print("Shape:", tensor_3d.shape)
print("Dimensions:", tensor_3d.ndim)
```

```python
print("Rank:", tensor_3d.ndim)
print("Shape:", tensor_3d.shape)
print("Size:", tensor_3d.size)

```

```python
students = np.array([
    [8.5, 110, 1],   # Student 1
    [7.2, 102, 0],   # Student 2
    [9.0, 120, 1]    # Student 3
])

print(students)
print("Shape:", students.shape)
print("Dimensions:", students.ndim)
```

```python
placement = np.array([1, 0, 1])

print(placement)
print("Shape:", placement.shape)
```

```python
# Vocabulary
hi = np.array([1, 0, 0])
hello = np.array([0, 1, 0])
bye = np.array([0, 0, 1])

sentence = np.array([hi, hello])

print(sentence)
print("Shape:", sentence.shape)
```

```python
data = np.array([
    [[1,0,0], [0,1,0]],   # sentence 1
    [[0,1,0], [0,0,1]]    # sentence 2
])

print(data.shape)
```
